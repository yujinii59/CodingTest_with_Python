#### 문제
인체에 치명적인 바이러스를 연구하던 연구소에서 바이러스가 유출되었습니다. 다행히 바이러스는 아직 퍼지지 않았고, 바이러스의 확산을 막기 위해서 연구소에 벽을 세우려고 합니다.

연구소는 크기가 N x M 인 직사각형으로 나타낼 수 있으며, 직사각형은 1 x 1 크기의 정사각형으로 나누어져 있습니다. 연구소는 빈칸, 벽으로 이루어져 있으며, 벽은 칸 하나를 가득 차지합니다.

일부 칸은 바이러스가 존재하며, 이 바이러스는 상하좌우로 인접한 빈칸으로 모두 퍼져나갈 수 있습니다. 새로 새울 수 있는 벽의 개수는 3개이며, 꼭 3개를 세워야 합니다.

#### 입력 조건
- 첫째 줄에 지도의 세로 크기 N과 가로 크기 M이 주어집니다. (3 <= N,M <= 8)
- 둘째 줄부터 N개의 줄에 지도의 모양이 주어집니다. 0은 빈간, 1은 벽, 2는 바이러스가 있는 위치입니다. 2의 개수는 2보다 크거나 같고, 10보다 작거나 같은 자연수입니다.
- 빈칸의 개수는 3개 이상입니다.

#### 출력 조건

##### 예시 설명
예를 들어, 다음과 같이 연구소가 생긴 경우를 살펴보겠습니다.

이때, 0은 빈칸, 1은 벽, 2는 바이러스가 있는 곳입니다. 아무런 벽을 세우지 않는다면, 바이러스는 모든 빈칸으로 퍼져나갈 수 있습니다. 2행 1열, 1행 2열, 4행 6열에 벽을 세운다면 지도의 모양은 다음과 같아집니다.

0 1 1 0 1 2 2

벽을 3개 세운 뒤, 바이러스가 퍼질 수 없는 곳을 안전 영역이라고 할 때 위의 지도에서 안전 영역의 크기는 27입니다.

연구소의 지도가 주어졌을 때 얻을 수 있는 안전 영역 크기와 최댓값을 구하는 프로그램을 작성하세요.



```python
from itertools import combinations
from collections import deque
from copy import deepcopy
import numpy as np
```

```python
moves = [(-1, 0), (1, 0), (0, -1), (0, 1)]
def virus_spread(map_info, x, y, row, col):
    for move in moves:
        move_x = x + move[0]
        move_y = y + move[1]
        if move_x < 0 or move_x >= row or move_y < 0 or move_y >= col:
            continue
        
        if map_info[move_x][move_y] == 0:
            map_info[move_x][move_y] = 2
            map_info = virus_spread(map_info, move_x, move_y, row, col)
    return map_info       

n, m = map(int, input().split())
maps = []
blanks = []
virus = []
for i in range(n):
    infos = list(map(int, input().split()))
    maps.append(infos)
    for j in range(m):
        info = infos[j]
        if info == 0:
            blanks.append((i, j))
            
        elif info == 2:
            virus.append((i, j))
            
            
cases = combinations(blanks, 3)

max_safe = 0
for case in cases:
    tmp_maps = deepcopy(maps)
    for x, y in case:
        tmp_maps[x][y] = 1
    for virus_x, virus_y in virus:    
        tmp_maps = virus_spread(tmp_maps, virus_x, virus_y, n, m)
        
    cnt = 0
    for status in tmp_maps:
        cnt += sum(1 for s in status if s == 0)
    
    if max_safe < cnt:
        max_safe = cnt 
print(max_safe)
        
```
